```python
import matplotlib.pyplot as plt
import numpy as np

# Coordinate dei punti
x0, y0 = 2, 4
x1, y1 = 6, 1

# Calcolo dell'angolo theta in radianti
theta = np.arctan2(y1 - y0, x1 - x0)

# Punti per la retta passante per (x0, y0) e (x1, y1)
line_x = [x0, x1]
line_y = [y0, y1]

# Perpendicolare da (x0, y0) alla retta orizzontale
perp_x = [x0, x0]
perp_y = [y0, y1]

# Proiezione ortogonale
proj_x = x0
proj_y = y1

# Inizializza la figura
plt.figure(figsize=(8, 6))
plt.plot(line_x, line_y, 'k-')  # Linea principale
plt.plot(perp_x, perp_y, 'gray', linestyle='--')  # Perpendicolare
plt.plot([0, x1], [0, 0], 'k--')  # Asse x per riferimento

# Punti principali
plt.plot(x0, y0, 'ko')  # Punto (x0, y0)
plt.plot(x1, y1, 'ko')  # Punto (x1, y1)
plt.plot(proj_x, proj_y, 'ko')  # Proiezione

# Etichette dei punti
plt.text(x0 + 0.2, y0 + 0.2, r'$(x_0, y_0)$', fontsize=12)
plt.text(x1 + 0.2, y1, r'$(x_1, y_1)$', fontsize=12)
plt.text(proj_x - 0.8, proj_y + 0.2, r'$\perp$', fontsize=14)

# Angoli
plt.text(x0 - 0.7, y0 - 0.2, r'$\theta$', fontsize=14)
plt.text(0.5, 0.2, r'$\theta$', fontsize=14)
plt.text(1.3, 0.3, r'$\rho, \theta$', fontsize=12)

# Formula tangente
plt.text(0, 5, r'$\tan \bar{\theta} = \frac{x_0 - x_1}{y_1 - y_0}$', fontsize=12)

# Formula y - y0 = ...
plt.text(7.5, 2.5, r'$y - y_0 = \frac{y_1 - y_0}{x_1 - x_0} (x - x_0)$', fontsize=12)

# Formula rho
plt.text(7.5, 2, r'$\rho = \tilde{x} \cos \bar{\theta}$', fontsize=12)

# Coordinate nel nuovo sistema
plt.text(7.5, 1.5, r'$\tilde{\rho}, \bar{\theta}$', fontsize=12)
plt.text(7.5, 1, r'$x_0 \cos \bar{\theta} + y_0 \sin \bar{\theta}$', fontsize=12)

# Impostazioni del grafico
plt.axis('equal')
plt.grid(True)
plt.xlim(-1, 10)
plt.ylim(-1, 6)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Schema geometrico con angolo e proiezioni')

plt.show()
```
